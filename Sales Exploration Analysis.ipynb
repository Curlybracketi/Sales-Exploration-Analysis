```python
ax = df.query('Age_Group != "others"')\
    .groupby('Age_Group')['Customer_Age']\
    .agg(['mean', 'count'])\
    .query('count >=20')\
    .sort_values('mean')\
    .plot(kind='barh', figsize=(12, 5), title='Average Customer Group Age')
ax.set_xlabel('Average Age')
plt.show()
```
